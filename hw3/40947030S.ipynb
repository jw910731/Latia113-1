# 題一

```python
# Generate the DataFrame
dates = pd.date_range("20130101", periods=100)
df = pd.DataFrame(np.random.randn(100, 4), index=dates, columns=list("ABCD"))

# Calculate the mean for each column
df_mean = df.mean().abs()

# Combined bar chart for all columns
plt.figure(figsize=(8, 6))
df_mean.plot(kind='bar', color=['red', 'blue', 'green', 'purple'], alpha=0.7)
plt.title("Mean Values of Columns A, B, C, D")
plt.ylabel("Mean Value")
plt.xlabel("Columns")
plt.show()

# Combined pie chart for all columns
plt.figure(figsize=(8, 6))
df_mean_abs = df_mean.abs()  # Use absolute mean values for pie chart
df_mean_abs.plot(kind='pie', autopct='%1.1f%%', colors=['red', 'blue', 'green', 'purple'], startangle=90)
plt.title("Proportion of Mean Values in Columns A, B, C, D")
plt.ylabel("")
plt.show()

# Combined 3D bar chart for all columns
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
x_pos = np.arange(len(df_mean))
y_pos = [0] * len(df_mean)  # Fixed y position for all bars
z_pos = [0] * len(df_mean)  # Fixed z position for all bars
dx = dy = 0.5
dz = df_mean.values

ax.bar3d(x_pos, y_pos, z_pos, dx, dy, dz, color=['red', 'blue', 'green', 'purple'], alpha=0.7)
ax.set_xlabel("Columns")
ax.set_ylabel("Y-Axis (fixed)")
ax.set_zlabel("Mean Value")
ax.set_xticks(x_pos)
ax.set_xticklabels(df_mean.index)
ax.set_title("3D Bar Chart of Mean Values for Columns A, B, C, D")
plt.show()
```
