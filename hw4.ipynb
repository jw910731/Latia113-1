```python
import requests
import json

web = requests.get('https://data.kcg.gov.tw/dataset/6f29f6f4-2549-4473-aa90-bf60d10895dc/resource/30dfc2cf-17b5-4a40-8bb7c511ea166bd3/download/lightrailtraffic.json')

# Check if the request was successful
if web.status_code == 200:
    # If successful, decode and process the JSON data
    data = json.loads(web.content.decode('utf-8-sig'))  # 手动处理 UTF-8 BOM
    print(data)
else:
    # If unsuccessful, print the error message
    print(f"Error: Request failed with status code {web.status_code}")
    print(web.text)
```

```python
soup = BeautifulSoup(html, "html.parser")
for script in soup(["script", "style"]):
    script.extract() # Indented this line to be part of the for loop
    # get text
text = soup.get_text()
# create a new bs4 object from the html data loaded
# remove all javascript and stylesheet code
# break into lines and remove leading and trailing space on each
lines = (line.strip() for line in text.splitlines())
# break multi-headlines into a line each
chunks = (phrase.strip() for line in lines for phrase in line.split(" "))
# drop blank lines
texts = '\n'.join(chunk for chunk in chunks if chunk)
print(texts)
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ptt.cc/' # This line was indented, causing the error
web = requests.get('https://www.ptt.cc/bbs/Gossiping/index.html',
                   cookies={'over18':'1'})
soup = BeautifulSoup(web.text, "html.parser")
titles = soup.find_all('div', class_='title')
# 取得class 為title 的div 內容
for i in titles:
    if i.find('a') != None:                         # 判斷如果不為None
        print(i.find('a').get_text())              # 取得 div 裡 a 的內容，使用get_text() 取得文字
        print(url + i.find('a')['href'], end='\n\n')  # 使用 ['href'] 取得href 的屬性
```

```python
import requests
url = 'https://invoice.etax.nat.gov.tw/index.html' # Removed the extra space before 'url'
web = requests.get(url)    # 取得網頁內容
web.encoding='utf-8'
# 上.encoding 避免亂碼 # Fixed indentation and added '#' to make this line a comment
# 因為該網頁編碼為utf-8，加
from bs4 import BeautifulSoup
soup = BeautifulSoup(web.text, "html.parser")
# 轉換成標籤樹 # Fixed indentation and added '#' to make this line a comment
td = soup.select('.container-fluid')[0].select('.etw-tbiggest')  #
# 取出中獎號碼的位置 # Fixed indentation and added '#' to make this line a comment
ns = td[0].getText()  # 特別獎
n1 = td[1].getText()  # 特獎
# 頭獎，因為存入串列會出現/n 換行符，使用[-8:] 取出最後八碼
n2 = [td[2].getText()[-8:], td[3].getText()[-8:], td[4].getText()[-8:]]
print(ns)
print(n1)
print(n2)
```

```python
import requests

url = 'https://rate.bot.com.tw/xrt/flcsv/0/day'   # 牌告匯率 CSV 網址
rate = requests.get(url)   # 爬取網址內容
rate.encoding = 'utf-8'    # 調整回應訊息編碼為 utf-8，避免編碼不同
# 造成亂碼
rt = rate.text
rts = rt.split('\n')
for i in rts:
    try:
        a = i.split(',')
        # 以文字模式讀取內容
        # 使用「換行」將內容拆分成串列
        # 讀取串列的每個項目
        # 使用try 避開最後一行的空白行
        # 每個項目用逗號拆分成子串列
        print(a[0] + ': ' + a[12])   # 取出第一個 ( 0 ) 和第十三個項目 ( 12 )
    except:
        break
```
